```python
# Assign the dataframe values from the previous cell into a new dataframe variable
# and make a bar plot with the text values horizontally and the numeric values as the bar heights
theaters = df.loc[range(0,10),['Statistics.Capacity', 'Show.Theatre']]
theaters.plot(kind='bar', x='Show.Theatre', y='Statistics.Capacity')
```

```python
# Re-do the plot from the previous cell as a horizontal bar plot
theaters.plot(kind='barh', x='Show.Theatre', y='Statistics.Capacity')
```

```python
# Re-do the plot from the previous cell
# and change at least two aesthetic elements (colors, labels, titles, ...)
barcolors = []
for i in range(0,10):
    if i%2:
        barcolors.append('grey')
    else:
        barcolors.append('orange')

ax = theaters.plot(kind='barh', 
              x='Show.Theatre', 
              y='Statistics.Capacity',
              color=barcolors
             )
ax.set_title('Broadway Theatres Capacity')
ax.set_xlabel('Capacity',fontsize=14)
ax.set_ylabel('Theatre',fontsize=14)
```

```python
# Mathemtical operation
shows = df['Show.Name'].unique()
shows

# Mathemtical operation
total_attendance = {}
for index,row in df.iterrows():
    show = row['Show.Name']
    if show not in total_attendance:
        total_attendance[show] = 0
    total_attendance[show] += row['Statistics.Attendance']

att = pd.DataFrame.from_dict(total_attendance, orient='index')
att
```

```python
att.plot(kind='bar', figsize=(100,5))
```

```python
after_2010 = df[df['Date.Year'] >= 2010].groupby('Date.Year').count()
after_2010.plot(kind='bar', y='Statistics.Performances').set_title('Number of Broadway Performances every year after 2010')
```
