The dataset I've chosen contains data about Broadway shows, grouped over weeklong periods. Only shows that reported capacity were included, so the dataset stretches back to the 1990s. We can view the data online at http://www.broadwayleague.com/.

The Python module can be obtained from https://corgis-edu.github.io/corgis/python/broadway/


```python
# Importing module to read the datafile
import broadway

# production is a list of dictionaries that represent objects with the dataset attributes
production = broadway.get_production()
print("The dataset size is", len(production))
```

#### Importing a module and using it to pretty print the data

```python
# module import and use of module import
from pprint import pprint as prettyprint

# Pretty printing the first object in the dataset
prettyprint(production[0])
```

### We start to explore the dataset features.

```python
print("The attributes of the datasets contains:", production[0].keys())

print()
print("Date's attributes:", production[0]['Date'].keys())
print("Show's attributes:", production[0]['Show'].keys())
print("Statistics's attributes:", production[0]['Statistics'].keys())
```

#### Using mathematical operation to calculate total number of attendees in the dataset 

```python
# Mathemtical operation
total_attendance = 0
for i in production:
    total_attendance += i['Statistics']['Attendance']
print("The Broadway dataset has a total of " + str(total_attendance) + " attendees!")
```

#### Using conditional expression to count the number of shows occurred after 2010

```python
# Conditional Expression
total = 0
for i in production:
    if i['Date']['Year'] > 2010:
        total += 1
print("There are total of " + str(total) + " Broadway shows in this dataset that occurred after 2010!")
```

#### Using a function to calculate the average number of performances for all the shows in the dataset

```python
# Function
def calcAvgNumOfPerformances(dataset):
    datasetLen = len(dataset)
    totalPerformances = 0
    for i in dataset:
        totalPerformances += i['Statistics']['Performances']
    return totalPerformances / datasetLen

avg = calcAvgNumOfPerformances(production)
print("The average number of performances in this dataset is", avg)
```

#### Using a for loop to print the date and attendance of each show in the dataset

```python
# For loop
for item in production:
    print(str(item['Date']['Full']) + " : " + str(item['Show']['Name']) + " has attendance " + str(item['Statistics']['Attendance']))
```
